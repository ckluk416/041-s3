```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# inisiasi graf G1
G1 = nx.Graph()

# menambahkan nodes
nodes_G1 = ['a', 'b', 'c', 'd']
G1.add_nodes_from(nodes_G1)

print(G1)

edges_G1 = [('a', 'b'), ('a', 'c'), ('a', 'd'), ('b', 'c'), ('c', 'd')]
G1.add_edges_from(edges_G1)

print(G1)

# Menghitung dan menampilkan properti graf
print("--- Analisis Graf G1 ---")
print(f"Jumlah nodes: {G1.number_of_nodes()}")
print(f"Jumlah edges: {G1.number_of_edges()}")
print("Derajat setiap node:")
for node in G1.nodes():
    print(f"  - Nodes '{node}': {G1.degree[node]}")

# Memvisualisasikan graf
plt.figure(figsize=(6, 6))
pos = {'a': (0, 1), 'b': (1, 1), 'c': (1, 0), 'd': (0, 0)} # Posisi simpul agar rapi
nx.draw(G1, pos=pos, with_labels=True, node_color='lightblue', font_weight='bold', font_size=12, node_size=700)
plt.title("Visualisasi Graf Sederhana G1")
plt.show()
```

```python
# Membuat objek multigraf
G2 = nx.MultiGraph()

# Menambahkan simpul dan sisi
G2.add_nodes_from(['a', 'b', 'c', 'd'])
edges_g2 = [('a', 'b'), ('a', 'c'), ('a', 'd'), ('c', 'd'), ('c', 'd'), ('a', 'd'), ('a', 'd'), ('b','c')]
G2.add_edges_from(edges_g2)

nx.draw(G2, with_labels=True, node_color='lightgreen', font_weight='bold')
plt.title("Graf Ganda G2")
plt.show()
```

```python
# Membuat objek graf semu (multigraf)
G3 = nx.MultiGraph()

# Menambahkan simpul-simpul
G3.add_nodes_from(['a', 'b', 'c', 'd'])

# Menambahkan sisi-sisi. Sisi ganda dan gelang bisa ditambahkan.
edges_G3 = [('a', 'b'), ('a', 'd'), ('a', 'd'), ('b', 'c'), ('c', 'c'), ('c', 'd'), ('c', 'd')]
G3.add_edges_from(edges_G3)

# Menghitung dan menampilkan properti graf
print("--- Analisis Graf Semu G3 ---")
print(f"Jumlah simpul: {G3.number_of_nodes()}")
print(f"Jumlah sisi: {G3.number_of_edges()}")
print("Derajat setiap simpul:")
for node in G3.nodes():
    # Perhatikan bahwa NetworkX secara otomatis menghitung loop dua kali untuk derajat
    print(f"  - Simpul '{node}': {G3.degree[node]}")

# Memvisualisasikan graf
plt.figure(figsize=(7, 7))
pos = nx.spring_layout(G3) # Layout otomatis untuk visualisasi yang lebih baik
nx.draw(G3, pos=pos, with_labels=True, node_color='lightgreen', font_weight='bold', font_size=12, node_size=700)
plt.title("Visualisasi Graf Semu G3")
plt.show()


# Memvisualisasikan graf dengan sisi yang melengkung
plt.figure(figsize=(7, 7))
pos = {'a': (0, 0), 'b': (0, 1), 'c': (1, 1), 'd': (1, 0)} # Posisi simpul agar lebih jelas
nx.draw_networkx_nodes(G3, pos, node_color='lightgreen', node_size=700)
nx.draw_networkx_labels(G3, pos, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G3, pos, edgelist=[('a', 'b'), ('b', 'c')], width=2.0, edge_color='black')
nx.draw_networkx_edges(G3, pos, edgelist=[('c', 'c')], width=2.0, edge_color='black', connectionstyle="arc3,rad=0.3")
# Menggambar sisi ganda dengan lengkungan berbeda
nx.draw_networkx_edges(G3, pos, edgelist=[('a', 'd')], width=2.0, edge_color='black', connectionstyle="arc3,rad=0.2")
nx.draw_networkx_edges(G3, pos, edgelist=[('a', 'd')], width=2.0, edge_color='black', connectionstyle="arc3,rad=-0.2")
nx.draw_networkx_edges(G3, pos, edgelist=[('c', 'd')], width=2.0, edge_color='black', connectionstyle="arc3,rad=0.2")
nx.draw_networkx_edges(G3, pos, edgelist=[('c', 'd')], width=2.0, edge_color='black', connectionstyle="arc3,rad=-0.2")

plt.title("Visualisasi Graf Semu G3")
plt.show()
```

**Figure 1: A computer network**

```python
G1 = nx.Graph()
edges_G1 = [
    ("San Francisco", "Los Angeles"),
    ("San Francisco", "Denver"),
    ("Los Angeles", "Denver"),
    ("Denver", "Chicago"),
    ("Chicago", "Detroit"),
    ("Chicago", "New York"),
    ("Chicago", "Washington"),
    ("Detroit", "New York"),
    ("New York", "Washington")
]
G1.add_edges_from(edges_G1)

plt.figure(figsize=(8, 5))
nx.draw(G1, with_labels=True, node_color='lightblue', node_size=700, font_weight='bold')
plt.title("Figure 1 - A Computer Network")
plt.show()
```

**Figure 15: Single-elimination tournament**

```python
G6 = nx.DiGraph()
edges_G6 = [
    ("Georgia", "Stanford"), ("Gonzaga", "Xavier"),
    ("Notre Dame", "Oklahoma"), ("Nebraska", "Kentucky"),
    ("Iowa State", "Connecticut"), ("Mississippi State", "Florida State"),
    ("Tennessee", "Baylor"), ("San Diego State", "Duke"),

    # round 2
    ("Xavier", "Stanford"), ("Kentucky", "Oklahoma"),
    ("Florida State", "Connecticut"), ("Duke", "Baylor"),

    # round 3
    ("Oklahoma", "Stanford"), ("Baylor", "Connecticut"),

    # semifinal
    ("Stanford", "Connecticut")
]
G6.add_edges_from(edges_G6)

plt.figure(figsize=(12, 6))
nx.draw(G6, with_labels=True, node_color='lightblue', node_size=700, font_weight='bold', arrows=True)
plt.title("Figure 15 - Single-Elimination Tournament")
plt.show()
```
